```julia
using JuMP, GLPK
using CSV
```

```julia
distance_matrix = CSV.read("distance_matrix.csv", header=0);
distance_matrix = convert(Matrix, distance_matrix);
```

```julia
neighborhood_matrix = CSV.read("neighborhood_matrix.csv", header=1);
neighborhood_matrix = convert(Matrix,neighborhood_matrix);
```

```julia
store_matrix = CSV.read("store_matrix.csv", header=1);
store_matrix = convert(Matrix, store_matrix);
```

```julia
d_betas = CSV.read("d_betas.csv", header=0);
d_betas = convert(Matrix, d_betas);
```

```julia
l_betas = CSV.read("l_betas.csv", header=0);
l_betas = convert(Matrix, l_betas);
```

```julia
# variable_C: binary indicator of each of the 1000 random coordinates; [1000, 1] column vector

# D: d_betas * store_matrix^(T) * disrtance_matrix * variable_C

# L: l_betas * neighborhood_matrix^(T) * variable_N
```

```julia
using JuMP, GLPK

n_coords = size(distance_matrix[1,:])[1]

model = Model(with_optimizer(GLPK.Optimizer))

@variable(model, x[1:n_coords]>=0, Bin)

@constraint(model, sum(x[1:n_coords]) == 1)
#@constraint(model, transpose(d_betas) * transpose(store_matrix) * distance_matrix * x[1:n_coords] == D)
#@constraint(model, transpose(l_betas) * transpose(neighborhood_matrix) * x[1:n_coords] == L)

@objective(model, Max, sum(transpose(d_betas) * transpose(store_matrix) * distance_matrix * x[1:n_coords]) + sum(transpose(l_betas) * transpose(neighborhood_matrix) * x[1:n_coords]))

println(model)
```

```julia
@time begin
    stats = optimize!(model)
end

println("\nObjective Value: ", JuMP.objective_value(model), "\n")

varvals = JuMP.value.(x)

opt_coord = findall(x->x==1,varvals)
println("Optimal Coordinate at Index: ", opt_coord)

```
